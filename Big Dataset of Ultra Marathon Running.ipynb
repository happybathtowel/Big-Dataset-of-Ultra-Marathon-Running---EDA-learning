```python
# naming the USA races as "dataframe 2"
df2 = df[(df['Event distance/length'].isin(['50km','50mi'])) & (df['Year of event'] == 2020) & (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA') ]
```

```python
df2.shape
```

```python
# removing the (USA) tag from the Event name column
df2['Event name'].str.split('(').str.get(0)
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

```python
# clean up athlete age *by using the event year 2020 and subtracting the "Athlete year of birth" column value
df2['athelete_age'] = 2020 - df2['Athlete year of birth']
```

```python
# removing h from athlete performance column
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
# checking "h" is gone from the Athlete performance column
df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head(5)
```

```python
# Dropping columns we don't need:
# Athlete club
# Athlete country
# Athlete year of birth
# Athlete age category
df2 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis=1)
```

```python
# checking above code worked
df2.head(5)
```

```python
# clean up null values
```

```python
# checking where and how many null values are there
df2.isna().sum()
```

```python
df2[df2['athelete_age'].isna() == 1]
```

```python
# dropping the nas in athlete_age
df2 = df2.dropna()
```

```python
# checking to see if the null values in athlete_age is now gone:
df2[df2['athelete_age'].isna() == 1]
```

```python
# checking for duplicates:answer will show there are no hits = no duplicates are there.
df2[df2.duplicated() == True]
```

```python
# resetting index
df2.reset_index(drop = True)
```

```python
# fix types
# Step 1 figuring out the data types of each column
df2.dtypes
```

```python
df2['athelete_age'] = df2['athelete_age'].astype(int)
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
# checking the result
df2.dtypes
```

```python
#Rename columns so there are no more spaces
# Year of event                  int64
# Event dates                   object
# Event name                    object
# Event dates       object
# Event number of finishers      int64
# Athlete performance           object
# Athlete gender                object
# Athlete average speed        float64
# Athlete ID                     int64
# athelete_age                   int64
# dtype: object
df2 = df2.rename(columns = {'Year of event': 'year',
                            'Event dates' : 'race_day',
                            'Event name' : 'race_name',
                            'Event distance/length' : 'race_length',
                            'Event number of finishers' : 'race_number_of_finishers',
                            'Athlete performance' : 'athlete_performance',
                            'Athlete gender' : 'athlete_gender',
                            'Athlete average speed' : 'athlete_average_speed',
                            'Athlete ID' : 'athlete_id',
})





```
